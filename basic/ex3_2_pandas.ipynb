```python
def getData(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return '{}-{}-{}'.format(year, month, day)
getData(12345667)
```

```python
df['사용일자'].apply(getData)
```

```python
df['사용일자1'] = df['사용일자'].apply(getData)
df['사용일자1'].astype('datetime64')
```
