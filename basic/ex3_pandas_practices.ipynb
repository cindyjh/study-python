```python
# 17. 컬럼명을 Date는 '날짜'로, 'High'는 '고가'로, 'Low'는 '저가'로 변경
df.rename({'Date':'날짜', 'High':'고가', 'Low': '저가'}, axis='columns')
```

```python
# 18. 날짜('Date') 컬럼 삭제
df.drop(['Date'], axis=1)
```

```python
# 19. diff' 컬럼을 생성하고, 종가('Close')와 시가('Open')의 차이를 계산하여 채우기
df['diff'] = df['Close'] - df['Open']
df
```

```python
# 20. 'Date'  필드에 저장된 값을 파싱하여 'Year', 'Month', 'Day' 필드에 저장하기
df['temp'] = df['Date'].astype('datetime64[ns]')
df['Year'], df['Month'], df['Day'] = [
    [day.year for day in df['temp']],
    [day.month for day in df['temp']],
    [day.day for day in df['temp']],
]
df.drop(['temp'], axis=1)
```

```python
# 21. 모든 데이터를 소수점 둘째자리까지 표시하기
format = lambda x: "%.2f" % x if isinstance(x, (int, float, complex)) and not isinstance(x, bool) else x
df.applymap(format, na_action='ignore')
```

```python
# 22. 거래량 필드를 생성하고 세 자릿수마다 콤마(,)가 들어가도록 변경하기
format_in_thousands = lambda x: '{:,}'.format(x) if isinstance(x, (int, float, complex)) and not isinstance(x, bool) else x
df['거래량'] = df['Volume'].apply(format_in_thousands)
df
```

```python
# 23. 주가 데이터를 월('Month')별로 그룹핑하고, 거래량('Volumn')의 평균값을 표시하기
df.groupby(by='Month')['Volume'].mean().apply(lambda x: '%f' % x)
```

```python
# 24. 결측치가 있는지 확인하고, 결측치가 있을 경우 결측치 없애기
df.isna().sum()
df.dropna()
```

```python
# 25. 6월의 데이터와 8월의 데이터를 별도의 데이터프레임(dec_df, nov_df)으로 만들고 두 데이터프레임을 하나의 데이터프레임으로 이어붙이기
jun_df = df[df['Month'] == 6]
aug_df = df[df['Month'] == 8]
pd.concat([jun_df, aug_df])
```
